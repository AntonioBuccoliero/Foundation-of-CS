```python
#We need to find a unique index, so we check if "movie_id" from credits table and "id" from movies tableare the same
def check_uniqueness(df, column):
    return len(df[column]) == len(df[column].drop_duplicates())
check_uniqueness(credits,'movie_id') and check_uniqueness(movies,'id')
```

```python
#the index is unique,than to simplify the search i'll rename the column from movies database
movies.rename(columns={'id':'movie_id'}, inplace=True)
```

```python
#As we can see, some dataframes of both datasets are written in json. To analyze that dataframes we need to extract the object
#using the following function
def extract_json(df, column):
    result = df[column].apply(pd.read_json) 
    movie_id = df['movie_id'] 
    for i in range(len(df)):
        result[i]['movie_id'] = movie_id[i] 
    return pd.concat(result.values).reset_index().drop("index", axis = 1) 
```

```python
#Create table from json field, named as the attribute and Deleted from dataset movies and credits
cast = extract_json(credits,'cast')

crew = extract_json(credits,'crew')

genres = extract_json(movies,'genres')

keywords = extract_json(movies,'keywords')

production_companies = extract_json(movies,'production_companies')

production_countries = extract_json(movies,'production_countries')

spoken_languages = extract_json(movies,'spoken_languages')


movies = movies.drop(['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages'],axis = 1)

credits = credits.drop(['cast','crew'], axis = 1)
```

4) Extract the domain of each homepage

```python
def domaine(hp):
    domains = []
    homepage = hp.split(" ")
    for i in range(len(homepage)):
        domains.append(homepage[i].split('/')[2].split('.')[-1])
    return str(domains)[1:-1].replace("\'", "")

movieshome["domains"]=movieshome["homepage"].apply(domaine)
movieshome[["title","homepage","domains"]]
```

```python
#which films has domain "it"
movieshome[["title","domains"]][movieshome["domains"]== "it"]
```

5) Extract a set of normalized* tables

*usefull to reduce data redundancy and improve data integrity.

```python
cast.drop(['name', 'gender' ], axis = 1, inplace = True)
cast_norm = cast.set_index('credit_id')

crew.drop(['name', 'gender' ], axis = 1, inplace = True)
crew_norm = crew.set_index('credit_id')

genres.rename(columns = {'id': 'genre_id'}, inplace = True)
genres['genre_id'] = genres['genre_id'].apply(int)
genres_name = genres[['genre_id','name']].drop_duplicates()
genres_name_norm = genres_name.set_index('genre_id').sort_index()
genres.drop('name',inplace = True, axis = 1)
genres_norm = genres.set_index(['genre_id','movie_id'])

keywords.rename(columns = {'id': 'keyword_id'}, inplace = True)
keywords['keyword_id'] = keywords['keyword_id'].apply(int)
keywords_name = keywords[['keyword_id','name']].drop_duplicates()
keywords_name_norm = keywords_name.set_index('keyword_id').sort_index()
keywords.drop('name',inplace = True, axis = 1)
keywords_norm = keywords.set_index(['keyword_id','movie_id'])

production_companies.rename(columns ={'id': 'company_id'}, inplace = True)
production_companies_name = production_companies[['company_id','name']].drop_duplicates()
production_companies_name_norm = production_companies_name.set_index('company_id').sort_index()
production_companies.drop('name',inplace = True, axis = 1)
production_companies_norm = production_companies.set_index(['company_id','movie_id'])

production_countries_name = production_countries[['iso_3166_1','name']].drop_duplicates()
production_countries_name_norm = production_countries_name.set_index('iso_3166_1').sort_index()
production_countries.drop('name',inplace = True, axis = 1)
production_countries_norm = production_countries.set_index(['iso_3166_1','movie_id'])

spoken_languages_name = spoken_languages[['iso_639_1','name']].drop_duplicates()
spoken_languages_name_norm = spoken_languages_name.set_index('iso_639_1').sort_index()
spoken_languages.drop('name',inplace = True, axis = 1)
spoken_languages_norm = spoken_languages.set_index(['iso_639_1','movie_id'])
```

10) For each company and decade, compute the overall revenue

```python
movies["year"] = movies["release_date"].str.extract("(.*)-.*-.*",expand=True)

def isdecade(year):
    return str(year)[:3] + '0'
movies['decade'] = movies['year'].apply(isdecade)
companies_id = production_companies.merge(right=movies[['movie_id','revenue','decade']], on='movie_id')
companies_name = production_companies_name.merge(right=companies_id[['company_id','revenue','decade']], on='company_id')
tot_revenue=companies_name.groupby(by =['company_id','name','decade'], as_index=False).sum()
tot_revenue
```

```python
#which companies/company reach a revenue betther then 20000000000 and in which decade?
tot_revenue[tot_revenue.revenue > 20000000000]
```

11) For each decade, compute the company with maximum revenue

```python
best_revenue10= tot_revenue.groupby(by =['decade'])[["revenue"]].apply(max, axis = 0).reset_index()
best_revenue = best_revenue10["revenue"].values
decade = best_revenue10["decade"].values

for i in range(len(best_revenue10)):
    print(tot_revenue[tot_revenue["decade"]==decade[i]][tot_revenue["revenue"]==best_revenue[i]],"\n") 
```
